```
#бидэнд хэрэгтэй модулуудыг оруулж ирэх
import nltk, re, string, collections
from nltk.util import ngrams 

#Дата болох файлаа оруулж ирэх
with open("/content/data.txt", "r", encoding='utf-8') as file:
    text = file.read()

#Дата файл маань зөв эсэхийг шалган эхний 1000 тэмдэгтүүдийг хэвлэе
text[0:1000]
```

```
# Энэ хэсэгт XML болон шинэ мөр таслал болон 
# бусад тэмдэгтээс ангижрах (Цэгийг харин үлдээх)
# Томоор бичсэн эсэх хамаарахгүй гэх мэт

#Бүх XML маркуудыг арилгах
text = re.sub('<.*>','',text)

# ENDOFARTICLE гэх тэкстээс салгах 
# ngram хийж буй дата маань Англи буюу Latin цагаан толгойгоор бичсэн тул ENDOFARTICLE
# хийж буй болно
text = re.sub('ENDOFARTICLE.','',text)

# Тэмдэгтүүдээс салгах (Цэгээс бусад нь!)
punctuationNoPeriod = "[" + re.sub("\.","",string.punctuation) + "]"
text = re.sub(punctuationNoPeriod, "", text)

# Харагдацыг шалгах
print(text)
text[0:1000]
```

```
# Үгсээ дангаар нь тасалж авах
tokenized = text.split()

# Бүх bi-gramийн жагсаах
esBigrams = ngrams(tokenized, 2)

print(tokenized)
```

```
# Корпуст байгаа бүх bi-gramийн давтамжыг олох
esBigramFreq = collections.Counter(esBigrams)

# Хамгийн түгээмэл орсон 10 үгсийг олох
esBigramFreq.most_common(10)
```

```
n = 4

ngrams = ngrams(tokenized, n)
for gram in ngrams:
    print(gram)
```
